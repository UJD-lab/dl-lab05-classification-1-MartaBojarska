## Zadanie 5

**a. Dlaczego zamrażamy bazę w fazie 1?**
Zamrażanie bazy (ang. *freezing*) chroni jej wyuczone już wagi (filtry rozpoznające podstawowe kształty i krawędzie) przed "zniszczeniem". Na początku treningu nowa, dodana przez nas warstwa klasyfikacyjna (głowa) jest inicjowana losowo, przez co generuje duże błędy. Trenując w pierwszej fazie wyłącznie tę nową warstwę, stabilizujemy ją i dopasowujemy do wyjścia z modelu bazowego.

**b. Co daje fine-tuning w fazie 2?**
Fine-tuning (dostrajanie) polega na odmrożeniu wybranych, głębszych warstw modelu i kontynuowaniu treningu z bardzo małym współczynnikiem uczenia (learning rate). Daje to sieci możliwość zaktualizowania i zaadaptowania swoich zaawansowanych filtrów do specyficznych, drobnych detali nowego zbioru danych (np. unikalnych tekstur na płatkach kwiatów), co pozwala wycisnąć z modelu maksymalną dokładność.

**c. Jaką przewagę ma MobileNetV2 wytrenowany na ImageNet nad siecią trenowaną od zera na tym datasecie?**
Zastosowanie modelu wytrenowanego na zbiorze ImageNet (Transfer Learning) daje ogromną przewagę czasową i obliczeniową. Sieć ta potrafi już rozpoznawać uniwersalne cechy wizualne. Trenowanie klasyfikatora od zera (tzw. *from scratch*) wymagałoby tysięcy zdjęć dla każdej z 102 klas kwiatów oraz bardzo długiego czasu treningu, aby sieć najpierw nauczyła się dostrzegać krawędzie i kolory, a dopiero potem odróżniać konkretne gatunki.
